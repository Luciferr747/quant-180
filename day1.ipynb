```python
# Day01_Hello.ipynb - Setup demo
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# create a synthetic price series: geometric Brownian motion-like
np.random.seed(42)
n = 252 * 2  # two years of daily data (approx)
dt = 1/252
mu = 0.08  # annual drift ~8%
sigma = 0.2 # annual vol 20%

# simulate log-returns & prices
eps = np.random.normal(0,1,size=n)
log_returns = (mu - 0.5*sigma**2)*dt + sigma * np.sqrt(dt) * eps
prices = 100 * np.exp(np.cumsum(log_returns))  # start at 100
dates = pd.date_range(end=pd.Timestamp.today(), periods=n, freq='B') # business days
series = pd.Series(prices, index=dates, name='Price')
```

```python
# compute simple returns and log returns
simple_returns = series.pct_change().dropna()
log_ret = np.log(series).diff().dropna()

# sample stats
mean_simple = simple_returns.mean()
var_simple = simple_returns.var(ddof=1)  # sample variance
mean_log = log_ret.mean()
var_log = log_ret.var(ddof=1)

print("Simple returns mean (daily):", mean_simple)
print("Simple returns variance (daily):", var_simple)
print("Log returns mean (daily):", mean_log)
print("Log returns variance (daily):", var_log)
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,8), sharex=True)
series.plot(ax=axes[0], title='Synthetic Price Series')
axes[0].set_ylabel('Price')

simple_returns.plot(ax=axes[1], title='Daily Simple Returns (synthetic)')
axes[1].axhline(0, color='k', linewidth=0.6, linestyle='--')
axes[1].set_ylabel('Return')
plt.tight_layout()
plt.show()
```

```python
# Observations
"""- We simulated a price series using geometric Brownian motion parameters (mu=8%, sigma=20%).
- Mean and variance above are daily values. To annualize:
  - annual_mean  mean_daily * 252
  - annual_vol  sqrt(var_daily * 252)
"""
```
