```python
# Lists in Python
prices = [100, 102, 105, 103, 107, 110]

# Accessing elements
first_price = prices[0]
last_price = prices[-1]

# Slicing
middle_slice = prices[1:4]

print("First:", first_price)
print("Last:", last_price)
print("Slice:", middle_slice)
```

```python
simple_returns = []
for i in range(1, len(prices)):
    ret = (prices[i] - prices[i-1]) / prices[i-1]
    simple_returns.append(ret)

simple_returns
```

```python
def compute_return(p_t, p_prev):
    return (p_t - p_prev) / p_prev

print(compute_return(105, 100))
```

```python
def compute_returns(price_list):
    returns = []
    for i in range(1, len(price_list)):
        r = compute_return(price_list[i], price_list[i-1])
        returns.append(r)
    return returns

compute_returns(prices)
```

```python
import numpy as np

prices_np = np.array(prices)
returns_np = (prices_np[1:] - prices_np[:-1]) / prices_np[:-1]

returns_np
```

```python
import numpy as np

prices_ex = [50, 51, 49, 52, 53]

def analyze_returns(price_list):
    # compute simple returns
    returns = [(price_list[i] - price_list[i-1]) / price_list[i-1] 
               for i in range(1, len(price_list))]
    
    max_ret = max(returns)
    min_ret = min(returns)
    avg_ret = sum(returns) / len(returns)
    
    return returns, max_ret, min_ret, avg_ret

returns, max_ret, min_ret, avg_ret = analyze_returns(prices_ex)

print("Returns:", returns)
print("Max return:", max_ret)
print("Min return:", min_ret)
print("Average return:", avg_ret)
```
