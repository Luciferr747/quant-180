```python
np.random.seed(42)

samples = np.random.normal(loc=0, scale=1, size=10000)

plt.figure(figsize=(10,5))
plt.hist(samples, bins=40, density= True, alpha=0.6)
plt.title("Standard Normal Distribution (10,000 samples)")
plt.show()
```

```python
sample_mean = samples.mean()
sample_var = samples.var(ddof=1)   # unbiased estimator

print("Sample mean:", sample_mean)
print("Sample variance:", sample_var)
```

```python
means = []
sizes = np.arange(10, 5000, 50)
for n in sizes:
    means.append(np.random.normal(0,1,size=n).mean())

plt.plot(sizes, means)
plt.axhline(0, color='red', linestyle='--')
plt.title("Law of Large Numbers")
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.show()
```

```python
means = []

N = 10000  # trials
for _ in range(N):
    sample = np.random.uniform(low=0, high=1, size=50)
    means.append(sample.mean())

plt.hist(means, bins=40, density=True, alpha=0.6)
plt.title("CLT: Means of Uniform Samples (n=50)")
plt.show()
```

```python
lognormal_samples = np.random.lognormal(mean=0, sigma=0.2, size=10000)

plt.hist(lognormal_samples, bins=40, density=True, alpha = 0.5)
plt.title("Lognormal Distribution  for Prices")
plt.show()
```

```python
from scipy.stats import t

df = 5
t_samples = t.rvs(df=df, size=10000)

plt.hist(t_samples, bins=40, density=True)
plt.title("Student-t Distribution (df=5)")
plt.show()
```

```python
from scipy.stats import norm

data = np.random.normal(0,1,1000)

# PDF
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.kdeplot(data, fill=True)
plt.title("Empirical PDF")

# CDF
plt.subplot(1,2,2)
sorted_data = np.sort(data)
cdf = np.arange(len(data)) / len(data)
plt.plot(sorted_data, cdf)
plt.title("Empirical CDF")

plt.show()
```
